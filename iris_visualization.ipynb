```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 读取数据
iris = load_iris()
iris_data = pd.DataFrame(iris['data'], columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])
iris_data['Species'] = iris.target
iris_data
```

```python
# 设置颜色主题
antV = ['#1890FF', '#2FC25B', '#FACC14', '#223273', '#8543E0', '#13C2C2', '#3436c7', '#F04864'] 
```

```python
# 绘制  Violinplot
import seaborn as sns
f, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True)
sns.despine(left=True)

sns.violinplot(x='Species', y='Sepal_Length', data=iris_data, palette=antV, ax=axes[0, 0])
sns.violinplot(x='Species', y='Sepal_Width', data=iris_data, palette=antV, ax=axes[0, 1])
sns.violinplot(x='Species', y='Petal_Length', data=iris_data, palette=antV, ax=axes[1, 0])
sns.violinplot(x='Species', y='Petal_Width', data=iris_data, palette=antV, ax=axes[1, 1])

plt.show()
```

```python
plt.style.use('seaborn')
sns.set_style("whitegrid")
```

```python
g = sns.lmplot(data=iris_data, x='Sepal_Width', y='Sepal_Length', palette=antV, hue='Species')
plt.plot()
```

```python

g2 = sns.lmplot(data=iris_data, x='Petal_Width', y='Petal_Length', palette=antV, hue='Species')
plt.plot()
```

```python
# 绘制  pointplot
f, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True)
sns.despine(left=True)

sns.pointplot(x='Species', y='Sepal_Length', data=iris_data, color=antV[1], ax=axes[0, 0])
sns.pointplot(x='Species', y='Sepal_Width', data=iris_data, color=antV[2], ax=axes[0, 1])
sns.pointplot(x='Species', y='Petal_Length', data=iris_data, color=antV[3], ax=axes[1, 0])
sns.pointplot(x='Species', y='Petal_Width', data=iris_data, color=antV[4], ax=axes[1, 1])

plt.show()
```

```python
# 各特征之间的关系
antV2 = ['#8543E0', '#F04864', '#FACC14', '#1890FF', '#2FC25B', '#223273', '#13C2C2', '#3436c7', ] 
g = sns.pairplot(data=iris_data, palette=antV2, hue= 'Species')
```
